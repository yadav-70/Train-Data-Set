 Cleaning and Preparing the data for model training

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the dataset
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
#Merging the both data frame 
df=df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
#Here we are using the map function map function is use to mapp the data using the give suitable conditions
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
#Handling the categorical feature Age
df['Age'].unique()
```

```python
# pd.get_dummies(df['Age'],drop_first=True)
df['Age'] = df['Age'].map({'0-17': 1, '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5, '51-55': 6, '55+': 8})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# LabelEncoder object to encode word labels
label_encoder = LabelEncoder()

# Encode labels in the 'Age' column
df['Age'] = label_encoder.fit_transform(df['Age'])

# Print unique values in the 'Age' column to verify encoding
df['Age'].unique()
```

```python
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
#Dronping th city Category 
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Focus on replacing missing values
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'].mode()[0]
```

```python
#Best way to replace the missing value that is Mode
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
#Product Category 3 replace missing values
df['Product_Category_3'].value_counts()
```

```python
#Best way to replace the missing value that is Mode
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['Product_Category_3'].isnull().sum()
```

```python
df.head()
```

```python
df.shape
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+', '', regex=False)
```

```python
df.head()
```

```python
df.info()
```

```python
#Covert object into the integer
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
df.info()
```

```python
#To converting the B & C 
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```
